```python
import pandas as pd
from scipy.stats import chi2_contingency

# 공공/민간 구분과 인적사고종류(대분류)의 교차 테이블 생성
contingency_table = pd.crosstab(df2['공공민간구분'], df2['인적사고종류(대분류)'])

# 카이제곱 검정 수행
chi2, p, dof, expected = chi2_contingency(contingency_table)

# 결과 출력
print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table:\n{expected}")

# 검정 결과 해석
alpha = 0.05
if p < alpha:
    print("두 변수는 통계적으로 유의미한 관계가 있습니다 (귀무가설 기각).")
else:
    print("두 변수는 통계적으로 유의미한 관계가 없습니다 (귀무가설 채택).")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 시각화할 사고 종류 리스트
accident_types = ['기타', '절단, 베임', '찔림', '넘어짐', '떨어짐', '화상']

# 필터링된 데이터프레임 생성
df_filtered = df2[df2['인적사고종류(대분류)'].isin(accident_types)]

# 각 사고 종류 쌍 간의 평균 차이를 시각화하는 함수
def plot_pairwise_comparisons(df, comparisons):
    plt.figure(figsize=(12, 8))
    sns.violinplot(x='온도', y='인적사고종류(대분류)', data=df, inner=None, palette="muted")
    
    for (acc1, acc2), p_adj, mean_diff in comparisons:
        data1 = df[df['인적사고종류(대분류)'] == acc1]['온도']
        data2 = df[df['인적사고종류(대분류)'] == acc2]['온도']
        plt.axvline(data1.mean(), color='blue', linestyle='--', label=f'{acc1} 평균')
        plt.axvline(data2.mean(), color='red', linestyle='--', label=f'{acc2} 평균')
        
        plt.text((data1.mean() + data2.mean()) / 2, 0.5, f'p-adj: {p_adj}\nMean diff: {mean_diff}',
                 ha='center', va='center', fontsize=12, color='black', bbox=dict(facecolor='white', alpha=0.5))
    
    plt.title('통계적으로 유의미한 사고 종류별 온도 분포 및 비교')
    plt.xlabel('온도')
    plt.ylabel('사고 종류')
    plt.legend()
    plt.show()

# 통계적으로 유의미한 쌍별 비교
comparisons = [
    (('기타', '절단, 베임'), 0.0351, 1.4438),
    (('기타', '찔림'), 0.0014, 3.3812),
    (('넘어짐', '떨어짐'), 0.0175, 0.9018),
    (('넘어짐', '찔림'), 0.0001, 3.61),
    (('찔림', '화상'), 0.039, -4.3482)
]

plot_pairwise_comparisons(df_filtered, comparisons)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 사고일시를 datetime 형식으로 변환
df2['사고일시'] = pd.to_datetime(df2['사고일시'])

# 월 파생변수 생성
df2['사고월'] = df2['사고일시'].dt.month

# 인적사고 종류(대분류)별 전체 사고 발생 건수 계산
total_accidents = df2['인적사고종류(대분류)'].value_counts()

# 상위 4개의 인적사고 종류(대분류) 선택
top_4_accidents = total_accidents.nlargest(4).index.tolist()

# 상위 4개의 사고 종류만 필터링
df_filtered = df2[df2['인적사고종류(대분류)'].isin(top_4_accidents)]

# 월별 상위 4개 인적사고 종류(대분류)별 사고 발생 빈도 계산
monthly_accidents_top4 = df_filtered.groupby(['사고월', '인적사고종류(대분류)']).size().reset_index(name='count')

# 선 그래프 그리기
plt.figure(figsize=(14, 8))
sns.lineplot(x='사고월', y='count', hue='인적사고종류(대분류)', data=monthly_accidents_top4, marker='o')

# 그래프 제목과 레이블 추가
plt.title('월별 상위 4개 인적사고 종류(대분류)별 사고 발생 빈도')
plt.xlabel('월')
plt.ylabel('사고 발생 건수')

# 범례 위치 조정
plt.legend(title='인적사고 종류(대분류)', loc='upper right')

# 그래프 표시
plt.show()
```
